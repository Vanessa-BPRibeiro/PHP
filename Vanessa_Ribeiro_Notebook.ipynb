```python
import numpy as np
import matplotlib.pyplot as plt

#aumentando o grafico
plt.figure(figsize=(10,5))

meses = ['Jul 21', 'Ago 21','Set 21', 'Out 21','Nov 21','Dez 21', 'Jan 22', 'Fev 22', 'Mar 22','Abr 22']
preco = [41553.7, 47130.4, 43823.3, 61309.6, 56882.9, 46219.5, 38498.6, 43188.2, 45525.0, 37650.0]

plt.bar(meses, preco, color='green')

plt.grid()

#legenda da barra no eixo X
plt.xticks(meses)

#label para o eixo y
plt.ylabel('Preço (USD)', color='blue')

#label para o eixo x
plt.xlabel('\nMeses', color='red')

#Titulo do grafico
plt.title('Variação do Bitcoin de Julho 2021 a Abril 2022 em dólares\n')

#mostrando o gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#tamanho do grafico
plt.figure(figsize=(15,10))

#Titulo do grafico
plt.title('Variação do Bitcoin de Julho 2021 a Abril 2022 em porcentagem\n\n')

meses = ['Jul 21', 'Ago 21','Set 21', 'Out 21','Nov 21','Dez 21', 'Jan 22', 'Fev 22', 'Mar 22','Abr 22']
preco = [41553.7, 47130.4, 43823.3, 61309.6, 56882.9, 46219.5, 38498.6, 43188.2, 45525.0, 37650.0]

#definindo fatias da pizza a serem destacadas das demais em suas respectivas posições
explode = (0, 0, 0, 0.2, 0, 0, 0, 0, 0, 0.2 )

colors=['gold', 'yellowgreen', 'coral', 'green','pink', 'grey', 'cyan','violet', 'lightskyblue', 'red']

plt.pie(preco, explode=explode, labels=meses, colors=colors, autopct='%1.0f%%', startangle=90)

plt.legend(preco, title='Preço(USD)\n', loc="lower right")

plt.axis('equal')

plt.show()
```
